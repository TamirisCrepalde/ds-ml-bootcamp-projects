## K Means Cluster Creation


```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters=2)
```

**fitting the model to all the data except for the Private label**

```python
kmean.fit(dt.drop('Private', axis=1))
```

**what are the cluster center vectors?**

```python
kmean.cluster_centers_
```

```python
kmean.labels_
```

___
## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels, however since this is just an exercise, we do have the labels, so we take advantage of this to evaluate our clusters, keep in mind, you usually won't have this luxury in the real world.


```python
def cluster(x):
    if x == 'Yes':
        return 1
    else:
        return 0
```

```python
dt['Cluster'] = dt['Private'].apply(cluster)
```

```python
dt.head()
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(dt['Cluster'], kmean.labels_))

print(classification_report(dt['Cluster'], kmean.labels_))
```
